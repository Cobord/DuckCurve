```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# Eyeballed from https://www.greentechmedia.com/articles/read/eia-charts-californias-real-and-growing-duck-curve#gs.yk3YFLk
# use accurate data if find it instead of just the graph image

data2012=[20,20,19,19,19,20,22,22,25,25,25,25,25,25,25,24,24,25,26,26,27,26,24,23]
data2013=[20,20,19,19,19,20,22,22,24,24,24,24,24,24,24,24,24,25,26,26,27,26,24,23]
data2014=[19,19,18,18,18,19,22,22,21,20,20,20,20,20,20,20,22,24,25,25,26,25,23,22]
data2015=[19,19,18,18,18,19,22,22,21,20,20,20,21,21,22,23,25,27,28,28,27,26,24,23]
data2016=[19,19,18,18,18,19,20,20,19,18,17,16,16,16,16,16,16,17,20,23,24,23,21,20]
data2017=[19,19,18,18,18,19,20,20,16,15,14,13,13,13,13,13,14,16,19,21,23,23,21,19]

data = [data2012,data2013,data2014,data2015,data2016,data2017]
X=[list(zip(np.ones(24)*(x+2012),np.arange(24))) for x in range(6)]
# rather than change the distance function on RBF just embed the circle into 2D plane
# distorts distances but just want a quick answer 
X=np.array([[item[0],np.cos(item[1]/12*np.pi),np.sin(item[1]/12*np.pi)] for sublist in X for item in sublist])
Y=np.array([item for sublist in data for item in sublist])
print(X[:10])
print(Y[:10])
print(Y.shape)
print(X.shape)
```

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF([10,2,2], (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gp.fit(X,Y)
xPred=[list(zip(np.ones(24)*(x+2012),np.arange(24))) for x in range(8)]
xPred=np.array([[item[0],np.cos(item[1]/12*np.pi),np.sin(item[1]/12*np.pi)] for sublist in xPred for item in sublist])
y_pred, sigma = gp.predict(xPred, return_std=True)
print(gp.get_params())

currentIndex=0
for i in range(8):
    fig = plt.figure()
    xPlot=np.arange(24)
    yPlot=y_pred[currentIndex:(currentIndex+24)]
    yPlotUB=y_pred[currentIndex:(currentIndex+24)]+1.9600*sigma[currentIndex:(currentIndex+24)]
    yPlotLB=y_pred[currentIndex:(currentIndex+24)]-1.9600*sigma[currentIndex:(currentIndex+24)]
    plt.plot(xPlot, yPlot, 'b-', label=u'Prediction'+str(i+2012))
    plt.fill(np.concatenate([xPlot, xPlot[::-1]]),
         np.concatenate([yPlotLB,
                        (yPlotUB)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval'+str(i+2012))
    plt.xlabel('hour')
    plt.ylabel('usage')
    plt.ylim(0, 30)
    if (i<6):
        plt.scatter(xPlot,Y[currentIndex:(currentIndex+24)],c='g')
    currentIndex+=24
```
